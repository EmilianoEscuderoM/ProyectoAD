Los estimados de locación nos ayudan a entender la tendencia central de los datos. La media nos da un promedio general, la mediana indica el punto medio de los datos y la moda muestra el valor que aparece con mayor frecuencia.
# Media

```python
# Calcular la media
mean_incidents = df_cleaned['Número_de_Incidentes'].mean()
mean_deaths = df_cleaned['Número_de_Muertes'].mean()
mean_displaced_people = df_cleaned['Número_de_Personas_Desplazadas'].mean()

print("Media:")
print(f"Número de Incidentes: {mean_incidents}")
print(f"Número de Muertes: {mean_deaths}")
print(f"Número de Personas Desplazadas: {mean_displaced_people}\n")
```

# Mediana


```python
# Calcular la mediana
median_incidents = df_cleaned['Número_de_Incidentes'].median()
median_deaths = df_cleaned['Número_de_Muertes'].median()
median_displaced_people = df_cleaned['Número_de_Personas_Desplazadas'].median()

print("Mediana:")
print(f"Número de Incidentes: {median_incidents}")
print(f"Número de Muertes: {median_deaths}")
print(f"Número de Personas Desplazadas: {median_displaced_people}\n")
```

# Moda

```python
# Calcular la moda
mode_incidents = df_cleaned['Número_de_Incidentes'].mode()[0]
mode_deaths = df_cleaned['Número_de_Muertes'].mode()[0]
mode_displaced_people = df_cleaned['Número_de_Personas_Desplazadas'].mode()[0]

print("Moda:")
print(f"Número de Incidentes: {mode_incidents}")
print(f"Número de Muertes: {mode_deaths}")
print(f"Número de Personas Desplazadas: {mode_displaced_people}")
```

# Calcular el rango intercuartílico (IQR)

```python
# Calcular el rango intercuartílico (IQR)
iqr_incidents = percentiles_incidents[0.75] - percentiles_incidents[0.25]
iqr_deaths = percentiles_deaths[0.75] - percentiles_deaths[0.25]
iqr_displaced_people = percentiles_displaced_people[0.75] - percentiles_displaced_people[0.25]

print("\nRango Intercuartílico (IQR):")
print(f"Número de Incidentes: {iqr_incidents:.2f}")
print(f"Número de Muertes: {iqr_deaths:.2f}")
print(f"Número de Personas Desplazadas: {iqr_displaced_people:.2f}")
```
